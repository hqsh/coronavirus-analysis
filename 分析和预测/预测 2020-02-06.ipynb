### 4.2. XGBoost 预测

predict 函数：取 2020-01-11 到 last_date 之间的各种数据，使用前 n - 1 天的数据训练并预测各地最后一天的新增确诊人数。

```python
from coronavirus_analyzer import CoronavirusAnalyzer
analyzer = CoronavirusAnalyzer('2020-02-07')  # '2020-02-03'
data = analyzer.get_data()
model = analyzer.get_model(data, print_log=True)
```

```python
from collections import OrderedDict
from coronavirus_analyzer import CoronavirusAnalyzer
import pandas as pd
import datetime
date = datetime.date(2020, 2, 1)
predicts = OrderedDict()
while date <= datetime.date.today():
    str_date = str(date)
    date += datetime.timedelta(days=1)
    analyzer = CoronavirusAnalyzer(str_date)
    data = analyzer.get_data()
#     model = analyzer.get_model(data, learning_rate=0.16, n_estimators=100, max_depth=4, min_child_weight=5, 
#                                gamma=0.4, subsample=1.0, colsample_bytree=1.0, reg_alpha=0, print_log=True)
    model = analyzer.get_model(data, print_log=True)
    s_X = data['df_X_test'].loc[str_date]
    predict = OrderedDict()
    for region in data['regions']:
        predict[region] = model.predict(s_X[region].values.reshape(1, -1))[0]
    predicts[str_date] = pd.Series(predict)
```

```python
import pandas as pd
import numpy as np
df_predict = pd.DataFrame(predicts).T
regions = df_predict.columns.tolist()
df_real = analyzer.df_virus_daily_inc_injured.loc[df_predict.index, df_predict.columns]
df_real = df_real.astype(np.int32)
df_predict.columns = pd.MultiIndex.from_product([df_predict.columns, ['预测']])
df_real.columns = pd.MultiIndex.from_product([df_real.columns, ['实际']])
df_predict = df_predict // 0.1 / 10
df = pd.concat([df_predict, df_real], axis=1).sort_index(axis=1)
df[regions]
```
