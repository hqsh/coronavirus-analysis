```python
from collections import OrderedDict
from coronavirus_analyzer import CoronavirusAnalyzer
import pandas as pd
import datetime
date = datetime.date(2020, 2, 1)
predicts = OrderedDict()
while date <= datetime.date.today():
    str_date = str(date)
    date += datetime.timedelta(days=1)
    analyzer = CoronavirusAnalyzer(str_date)
    data = analyzer.get_data()
    xgb_params = {'base_score': 0.5, 'booster': 'gbtree', 'colsample_bylevel': 1, 'colsample_bynode': 1, 
                  'colsample_bytree': 0.9, 'gamma': 0.3, 'importance_type': 'gain', 'learning_rate': 0.06, 
                  'max_delta_step': 0, 'max_depth': 5, 'min_child_weight': 5, 'missing': None, 
                  'n_estimators': 150, 'nthread': 1, 'objective': 'reg:gamma', 'reg_alpha': 0.0001, 
                  'reg_lambda': 1, 'scale_pos_weight': 1, 'seed': 0, 'subsample': 1.0, 'verbosity': 1}
    model = analyzer.get_model(data, print_log=True, **xgb_params)
    s_X = data['df_X_test'].loc[str_date]
    predict = OrderedDict()
    for region in data['regions']:
        predict[region] = model.predict(s_X[region].values.reshape(1, -1))[0]
    predicts[str_date] = pd.Series(predict)
```

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df_predict = pd.DataFrame(predicts).T
regions = df_predict.columns.tolist()
df_real = analyzer.df_virus_daily_inc_injured.loc[df_predict.index, df_predict.columns]
df_real = df_real.astype(np.int32)
df_predict.columns = pd.MultiIndex.from_product([['{}'.format(c) for c in df_predict.columns], ['预测']])
df_real.columns = pd.MultiIndex.from_product([df_real.columns, ['实际']])
df_predict = df_predict // 0.1 / 10
df = pd.concat([df_predict, df_real], axis=1)[regions]
df.to_excel('预测2020-02-07.xlsx')
df
```
